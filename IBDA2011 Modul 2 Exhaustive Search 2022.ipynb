## 2.1.1 Selection Sort

Salah satu pendekatan paling naif dalam menyelesaikan permasalahan Sorting adalah dengan secara satu per satu menentukan elemen hasil pengurutan mulai dari yang terkecil hingga terbesar. Gagasan sorting semacam ini sering disebut Selection Sort.

```
Input  : An array A of n elements
Output : Sorted version of A such that A[0] <= A[1] <= ... <= A[n-1]
```

Gambar besar strategi algoritma ini adalah sebagai berikut:

```
input(A)
for i = 0, 1, ... , n-1:
  A[j] := minimum element among {A[i], A[i+1], ... , A[n-1]}
  swap(A[i], A[j])
return A
```

````python
A = list(map(int, input().split(" ")))
n = len(A)
B = []

for i in range(n):
  min_element = A[i]
  min_idx = i
  for j in range(i+1, n):
    if A[j] < min_element:
      min_element = A[j]
      min_idx = j
  A[i], A[min_idx] = A[min_idx], A[i]

print(A)
````

Algoritma di atas jelas memiliki kompleksitas waktu dan memori yang bersifat kuadratik, yaitu $\mathcal{O}(n^2)$.

## 2.1.2 Perkalian Matriks

Masalah perkalian matriks merupakan salah satu masalah numerik yang paling klasik dan merupakan suatu latihan yang sangat baik untuk mengembangkan kemampuan berpikir abstrak seorang pemrogram pemula. Berikut merupakan deskripsi masalahnya.

```
Input  : Matrix A of size m x n and Matrix B of size n x t
Output : Matrix AB of size m x t, the result of multiplying A and B
```

Misalkan $A\in \mathbb{R}^{m\times n}$ dan $B\in \mathbb{R}^{n\times t}$. Misalkan pula $C=AB$, hasil perkalian kedua matriks. Maka $C\in \mathbb{R}^{m\times t}$.

Untuk dapat menghitung matriks $C$, maka tentu diperlukan perulangan sebanyak $m\times t$ kali untuk menghitung setiap elemen dari matriks $C$. Setiap elemen hasil perkalian matriks tersebut didefinisikan melalui formula berikut:

$$
c_{ij}=\sum_{k=1}^n a_{ik}b_{kj}, \qquad \text{untuk semua }i=1, 2, \dots , m; \quad j=1, 2, \dots, t.
$$

Maka, perhitungan perkalian matriks yang dilakukan secara naif akan menerapkan secara langsung formula di atas untuk setiap elemen matriks $C$.

````python
A = [
	[5, 3, 1],
	[0, -2, 4]
]
B = [
	[1, -1, 3],
	[2, 6, 0],
	[-2, -5, 1]
]
m = len(A)
n = len(A[0])
t = len(B[0])

C = [[0 for j in range(t)] for i in range(m)]

for i in range(m):
	for j in range(t):
    # compute the element C[i][j] of the i-th row and j-th column
		element = 0
		for k in range(n):
			element+= A[i][k] * B[k][j]
		C[i][j] = str(element)

for row in C:
	print("\t".join(row))
````

Mari kita ambil waktu sejenak untuk mengapresiasi hal ini.

Algoritma naif untuk menyelesaikan permasalahan ini sangat mirip dengan Sequential Search, yaitu mengecek satu demi satu setiap posisi string ```text``` kemudian menelusuri setiap digit string berikutnya apakah berisi sama dengan string ```pattern```, digit demi digit.

```
Input: An array T [0..n − 1] of n characters representing a text and an array P [0..m − 1] of m characters representing a pattern
Output: The index of the first character in the text that starts a matching substring or −1 if the search is unsuccessful
```

````python
T = input()
P = input()
n, m = (len(T), len(P))
found = True

for i in range(n-m):
  j = i
  found = True
  while j < i+m:
    if T[j]!=P[j-i]:
      found = False
      break
    j += 1
  if found:
    print(i)
    break
  
if not (found):
  print("-1")
````

Karena terdapat dua buah perulangan bersaran yang masing-maisng dilakukan sebanyak $n$ dan $m$ kali, maka kompleksitas waktu algoritma di atas adalah $\mathcal{O}(mn)$.

Sementara itu, kompleksitas memori yang dibutuhkan algoritma di atas hanyanya kapasitas untuk menyimpan dua buah string ```P``` dan ```T```, yaitu sebesar $\mathcal{O}(m+n)$.

Dengan demikian, permasalahan Convex Hull jika diselesaikan secara *Exhaustive Search* akan memiliki kerangka algoritma sebagai berikut.

```
Input  : Two lists x and y of n (n≥2) points p1(x1,y1),...,pn(xn,yn)
Output : The total length of the convex hull surrounding the points on the input

ConvexHull(x, y):
  ConvexHullLength = 0
  for i = 1, 2, ... , n:
    for j = i, i+1, ... , n:
      if IncludedInConvexHull(x[i], y[i], x[j], y[j]):
        ConvexHullLength += Distance(x[i], y[i], x[j], y[j])
  return ConvexHullLength
```

Bagaimana sekarang cara menentukan apakah suatu titik berada di sebelah kiri atau kanan dari suatu segmen garis? Pertanyaan ini menjadi kunci untuk mendesain fungsi ```IncludedInConvexHull()``` pada gambaran algoritma di atas.

Untuk menyelesaikan permasalahan ini, kita akan menggunakan sebuah lema geometri vektor sebagai berikut.

> **Lema.** Misalkan $\mathbf{\vec{AB}}$ dan $\mathbf{\vec{AC}}$ adalah dua vektor bebas linear di $\mathbb{R}^2$ yang memiliki titik mula yang sama, yaitu titik $A$. Misalkan $$\mathbf{\vec{AB}}=\begin{bmatrix} u_1\\u_2\end{bmatrix},\qquad \mathbf{\vec{AC}}=\begin{bmatrix} w_1\\w_2\end{bmatrix}. $$ Maka, vektor $\mathbf{\vec{AB}}$ berada di sebelah kanan vektor $\mathbf{\vec{AC}}$ jika dan hanya jika $$u_1w_2- u_2w_1>0.$$

> *Bukti:* Pandang vektor $\mathbf{\vec{AB}}$ dan $\mathbf{\vec{AC}}$ sebagai vektor di $\mathbb{R}^3$ yang terletak di bidang $xy$. Berdasarkan aturan tangan kanan, $\mathbf{\vec{AB}}$ berada di sebelah kanan $\mathbf{\vec{AC}}$ jika dan hanya jika vektor $\mathbf{\vec{AB}}\times \mathbf{\vec{AC}}$ mengarah ke atas (searah sumbu $z$ positif). Ini juga berarti bahwa $\lVert\mathbf{\vec{AB}}\times \mathbf{\vec{AC}}\rVert>0$. Perhatikan bahwa $\lVert\mathbf{\vec{AB}}\times \mathbf{\vec{AC}}\rVert=u_1w_2- u_2w_1$. Terbukti.

Lema di atas memberikan sebuah kenyamanan komputasi untuk mengecek apakah suatu titik terletak di sebelah kiri atau kanan dari sebuah segmen garis.

# 2.2 Exhaustive Search

Ingat kembali bahwa algoritma naif berfokus pada pendekatan penyelesaian masalah yang bersifat straightforward dan 'langsung'. Walaupun dengan semangat yang sama, Algoritma Exhaustive Search memiliki tambahan ciri khas, yaitu bahwa algoritma membangkitkan seluruh kemungkinan solusi yang mungkin terjadi untuk diperiksa satu demi satu.

Untuk memberi gambaran, perhatikan contoh masalah di bawah ini.

Misalkan kita hendak menjelajah sebuah array ```A``` berukuran ```N``` yang setiap petaknya berisi harta karun (bilangan bulat positif) atau sebuah ranjau (berisi bilangan bulat negatif). Berikut merupakan aturan untuk mengumpulkan harta karun:
* Kita selalu mulai dari posisi indeks 0, yaitu petak yang paling kiri.
* Di setiap langkah, kita dapat berpindah ke petak yang berjarak 1 atau berjarak 2 di sebelah kanan petak sebelumnya.

Berapa banyakkah total harta karun terbesar yang dapat diperoleh?

Sebagai contoh, misalkan ```N=4``` dan berikut merupakan isi array ```A```:

|   ```i```  |  0  |  1 | 2 | 3 |  4 |
|:----:|:---:|:--:|:-:|:-:|:--:|
| ```A[i]``` | -10 | -4 | 5 | 1 | -2 |

Untuk mengetahui rute pengumpulan harta karun terbesar, salah satu cara yang paling 'naif' adalah dengan menuliskan seluruh kemungkinan rute perjalanan yang mungkin ditempuh, kemudian membandingkannya satu demi satu untuk dicari rute mana yang menghasilkan harta karun terbesar. Berikut merupakan tabel yang mendaftarkan seluruh kemungkinan rute beserta hasil harta karun yang diperoleh pada masing-masing rute:

| No |       Rute Kunjungan (indeks)      |          Total Harta Karun          |
|:--:|:----------------------------------:|:-----------------------------------:|
|  1 | ```0 -> 1 -> 2 -> 3 -> 4 -> out``` | ```10 + (-4) + 5 + 1 + (-2) = 10``` |
|  2 |    ```0 -> 1 -> 2 -> 3 -> out```   |        ```10 + (-4) + 5 + 1 = 12``` |
|  3 |    ```0 -> 1 -> 2 -> 4 -> out```   |      ```10 + (-4) + 5 + (-2) = 9``` |
|  4 |    ```0 -> 1 -> 3 -> 4 -> out```   |      ```10 + (-4) + 1 + (-2) = 5``` |
|  5 |    ```0 -> 2 -> 3 -> 4 -> out```   |        ```10 + 5 + 1 + (-2) = 15``` |
|  6 |      ```0 -> 1 -> 3 -> out```      |             ```10 + (-4) + 1 = 6``` |
|  7 |      ```0 -> 2 -> 3 -> out```      |               ```10 + 5 + 1 = 16``` |
|  8 |      ```0 -> 2 -> 4 -> out```      |            ```10 + 5 + (-2) = 13``` |

Perhatikan dari 8 kemungkinan rute tersebut, rute ke 7 yang menghasilkan total harta karun terbesar, yaitu sebanyak 16. Metode mencari semua kemungkinan seperti di atas tentu pasti dengan cermat dapat menemukan solusi yang paling optimal, namun belum tentu dapat dilakukan dengan cepat. Inilah yang disebut sebagai strategi Complete Search.

Algoritma Complete Search kemudian berfokus pada teknik-teknik untuk membangkitkan semua kemungkinan solusi secara lengkap. Beberapa teknik yang umum digunakan adalah Algoritma DFS (Depth First Search) dan BFS (Breadth First Search).

Dalam konteks membangkitkan seluruh kemungkinan solusi, seringkali ruang solusi dapat dibangkitkan semua elemennya dengan cara 'dijelajah' langkah demi langkahnya. Untuk menyimpan informasi mengenai langkah penjelajahan ini, seringkali struktur data yang efektif adalah Graf atau Pohon.

## 2.2.1 Algoritma Depth First Search

Algoritma DFS berfokus pada penjelajahan yang bersifat 'mendalam' dibandingkan secara 'melebar'. DFS akan lebih memprioritaskan untuk segera menuntaskan seluruh langkah hingga ditemukan sebuah kandidat solusi ketimbang menyimpan seluruh kemungkinan solusi dan mengunjungi masing-masing kemungkinan secara langkah demi langkah.

Dalam permasalahan di atas, strategi algoritma DFS akan terus berupaya untuk mengunjungi titik dengan rute perjalanan dengan sisi biru secara murni terlebih dahulu hingga hal tersebut tidak dimungkinkan, baru mempertimbangkan untuk menggunakan kemungkinan lain, yaitu sisi warna merah. Artinya, dalam konteks pohon, DFS akan berangkat dari akar dan memprioritaskan untuk menjelajah hingga ke titik daun terlebih dahulu ketimbang meninjau alternatif penjelajahan dengan pilihan anak yang lain.

Secara umum, terdapat dua pendekatan untuk mengimplementasi DFS, yaitu secara rekursif dan dengan menggunakan struktur data *stack*.

### DFS Rekursif

Secara kasar, prosedur DFS dapat digambarkan secara rekursif sebagai berikut.

```
DFS(current_state, path_explored):
  
  # from current_state, identify all possible next steps
  possible_next_steps = <all next possible states attained from current_state>
  
  if possible_next_steps is empty then:
    # in this condition, we have found a complete solution candidate
    print(path_explored)
    return
  
  # if there is another possible next step within the exploration, then do the following  
  for next_state in possible_next_steps:
    # do the same exploration procedure from next_state
    DFS(next_state, path_explored + next_state)
```

Selain mendesain isi algoritma rekursif di atas, kita juga perlu dengan hati-hati menuliskan perintah pemanggilan awal prosedur dengan tepat. Secara umum algoritma di atas ketika dipanggil pertama kali perlu untuk memilih langkah awal yang tepat serta mengisikan sebuah path kosong dalam argumen fungsi. Berikut merupakan pemanggilan awal yang tepat setelah fungsi didefinisikan.

```
DFS(initial_state, empty_path)
```

Penyelesaian masalah di atas dengan menggunakan DFS Rekursif dapat ditinjau melalui potongan kode berikut.

````python
#@title DFS Rekursif  {display-mode: "form"}

A = [10, -4,	5,	1,	-2]
n = len(A)
all_solutions = []

def DFS(current_position, path_explored, n, current_val):
  # the procedure keeps track all the exploration paths
  # when an exploration ends, it produces a solution candidate
  # and is recorded in the list all_solutions
  
  if current_position >= n:
    # this is the case when the exploration finishes
    all_solutions.append([current_val, path_explored])
    return
  
  # this is the case when the exploration had not finished yet
  
  # now we explore current_position
  next_path_explored = path_explored + [current_position]
  current_val += A[current_position]

  # perform the exploration 1 step to the right
  next_position1 = current_position + 1
  DFS(next_position1, next_path_explored, n, current_val)

  # if the previous exploration would have reached the end,
  # then the next one is not necessary to conduct
  if next_position1 < n:
    # perform the exploration 2 step to the right
    next_position2 = current_position + 2
    DFS(next_position2, next_path_explored, n, current_val)

# initially the initial position is 0 and the path is empty
DFS(0, [], n, 0)

max_val = all_solutions[0][0]
for val, route in all_solutions:
  print("Rute: ", ' -> '.join(list(map(str,route))), "-> Finish")
  print("Total Harta Karun: ", val)
  if val > max_val:
    max_val = val
    best_route = route

print("\nJadi, rute terbaik adalah: ", ' -> '.join(list(map(str,best_route))), "-> Finish")
print("dengan total harta karun sebesar ", max_val)
````

### DFS dengan *Stack*

Selain dengan rekursif, DFS dapat dijalankan dengan menggunakan struktur data stack yang memiliki sifat *LIFO (Last In First Out)*. Ketika menjelajah, setiap pilihan langkah selanjutnya yang memungkinkan akan disimpan ke dalam stack untuk akhirnya satu demi satu pilihan langkah tersebut kemudian akan dieksekusi. Dengan sistem LIFO, algoritma DFS akan selalu memprioritaskan pilihan alternatif langkah yang paling terkini. Ini berarti bahwa penjelajahan pohon akan selalu memprioritaskan untuk bergerak menjelajahi setiap keturunan hingga menemukan daun terlebih dahulu sebelum menjelajah kemungkinan cabang lainnya.

Secara kasar, anatomi algoritma DFS dengan menggunakan Stack adalah sebagai berikut

```
def DFS(initial_state, empty_path):
  
  S = <empty stack>
  # push the initial state and empty path into the stack
  S.push([initial_state, empy_path])

  # we read all possible moves stored in the stack until there is no other further way to explore
  while not S.empty():
    # the latest state pushed on the stack is being read and we continue the exploration from there
    current_state, current_path = S.pop()
    possible_next_steps = <all next possible states attained from current_state>
    
    # from the current_state, explore all possible next steps and push them all into the stack
    for next_state in possible_next_steps:
      S.push([next_state, current_path + next_state]
```

Penyelesaian masalah di atas dengan menggunakan DFS dengan Stack dapat ditinjau melalui potongan kode berikut.

````python
#@title DFS dengan Stack  {display-mode: "form"}

from queue import LifoQueue

stack = LifoQueue()

A = [10, -4,	5,	1,	-2]
n = len(A)
all_solutions = []

def DFS(initial_position, initial_path, initial_value, n):
  # initially the initial position is 0 and the path is empty
  # put this state into the stack
  stack.put([initial_position, initial_path, initial_value])

  while not stack.empty():
    current_position, path_explored, current_val = stack.get()

    if current_position >= n:
      # this is the case when the exploration finishes
      all_solutions.append([current_val, path_explored])
    else:
      # this is the case when the exploration had not finished yet
  
      # now we explore current_position
      next_path_explored = path_explored + [current_position]
      current_val += A[current_position]

      # perform the exploration 1 step to the right
      next_position1 = current_position + 1
      stack.put([next_position1, next_path_explored, current_val])

      # if the previous exploration would have reached the end,
      # then the next one is not necessary to conduct
      if next_position1 < n:
        # perform the exploration 2 step to the right
        next_position2 = current_position + 2
        stack.put([next_position2, next_path_explored, current_val])

DFS(0, [], 0, n)

max_val = all_solutions[0][0]
for val, route in all_solutions:
  print("Rute: ", ' -> '.join(list(map(str,route))), "-> Finish")
  print("Total Harta Karun: ", val)
  if val > max_val:
    max_val = val
    best_route = route

print("\nJadi, rute terbaik adalah: ", ' -> '.join(list(map(str,best_route))), "-> Finish")
print("dengan total harta karun sebesar ", max_val)
````

## 2.2.2 Algoritma Breath First Search

Algoritma BFS berfokus pada penjelajahan yang bersifat 'melebar' dibandingkan secara 'mendalam'. BFS akan lebih memprioritaskan untuk menelusuri setiap langkah baru yang mungkin muncul ketimbang terburu-buru untuk menuntaskan seluruh langkah hingga ditemukan sebuah kandidat solusi.

Dalam permasalahan di atas, strategi algoritma BFS akan menjelajah setiap anak dari sebuah titik tanpa secara terburu-buru untuk menjelajah lebih dalam anak dari anak titik yang sedang ditinjau. Artinya, strategi ini akan bergerak secara 'rapi' mulai dari tingkat 0 pohon, yaitu langkah awal, menelusuri semua titik di tingkat 1, baru kemudian menelusuri titik di tingkat 2 dst. BFS tidak akan berpindah menelusuri tingkat yang lebih dalam jika masih ada titik lain yang masih setingkat dan belum dijelajahi. Dalam penelusuran, jika sekiranya diperoleh kandidat solusi (yaitu sebuah daun), maka algoritma BFS menyimpannya dan melanjutkan penelusuran seperti biasa tanpa adanya kemungkinan *backtracking*.

Kontras dengan DFS, struktur data yang tepat untuk digunakan dalam algoritma ini adalah ```Queue```. Secara kasar, anatomi algoritma DFS dengan menggunakan Stack adalah sebagai berikut

```
def BFS(initial_state, empty_path):
  
  Q = <empty queue>
  # push the initial state and empty path into the queue
  Q.push([initial_state, empy_path])

  # we read all possible moves stored in the queue until there is no other further way to explore
  while not Q.empty():
    # the latest state pushed on the queue is being read and we continue the exploration from there
    current_state, current_path = Q.pop()
    possible_next_steps = <all next possible states attained from current_state>
    
    # from the current_state, explore all possible next steps and push them all into the queue
    for next_state in possible_next_steps:
      Q.push([next_state, current_path + next_state]
```

Penyelesaian masalah di atas dengan menggunakan BFS dapat ditinjau melalui potongan kode berikut.

````python
#@title BFS  {display-mode: "form"}

from queue import Queue

queue = Queue()

A = [10, -4,	5,	1,	-2]
n = len(A)
all_solutions = []

def BFS(initial_position, initial_path, initial_value, n):
  # initially the initial position is 0 and the path is empty
  # put this state into the queue
  queue.put([initial_position, initial_path, initial_value])

  while not queue.empty():
    current_position, path_explored, current_val = queue.get()

    if current_position >= n:
      # this is the case when the exploration finishes
      all_solutions.append([current_val, path_explored])
    else:
      # this is the case when the exploration had not finished yet
  
      # now we explore current_position
      next_path_explored = path_explored + [current_position]
      current_val += A[current_position]

      # perform the exploration 1 step to the right
      next_position1 = current_position + 1
      queue.put([next_position1, next_path_explored, current_val])

      # if the previous exploration would have reached the end,
      # then the next one is not necessary to conduct
      if next_position1 < n:
        # perform the exploration 2 step to the right
        next_position2 = current_position + 2
        queue.put([next_position2, next_path_explored, current_val])

BFS(0, [], 0, n)

max_val = all_solutions[0][0]
for val, route in all_solutions:
  print("Rute: ", ' -> '.join(list(map(str,route))), "-> Finish")
  print("Total Harta Karun: ", val)
  if val > max_val:
    max_val = val
    best_route = route

print("\nJadi, rute terbaik adalah: ", ' -> '.join(list(map(str,best_route))), "-> Finish")
print("dengan total harta karun sebesar ", max_val)
````

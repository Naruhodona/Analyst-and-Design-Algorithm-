## 3.1 Decrease and Conquer

Dalam strategi ini, masalah diselesaikan dengan cara ukuran input yang terus dikecilkan secara bertahap agar pengerjaan lebih mudah dikerjakan. Tiga pendekatan utama yang digunakan dalam strategi ini antara lain:
1. *Decrease by a constant*
2. *Decrease by a constant factor*
3. *Variable size decrease*

Ketiganya memiliki cara kerja yang sama, yaitu menyelesaikan sebuah masalah dengan melakukan manipulasi tertentu agar input yang ukurannya besar dapat berkurang ukurannya hingga cukup kecil dan cenderung 'mudah' untuk diselesaikan.

### 3.1.1 Decrease by a constant

Dalam strategi ini, ukuran input sebuah masalah akan berkurang secara konstan dalam setiap iterasinya, seperti menyelesaikan fungsi rekursif pada normalnya. Perhatikan contoh sederhana di bawah.

$$
f(n)=
\begin{cases}
n f(n-1), & n>0\\
1, & n=0
\end{cases}
$$

Maka, untuk menghitung nilai $f(4)$ misalnya, maka kita akan terus memanggil fungsi $f$ dengan input yang lebih kecil secara iteratif hingga diperoleh nilai $n=0$, agar kita dapat mendapatkan sebuah nilai angka.

$$
\begin{align}
f(4) &= 4f(3)\\
& = 4.3.f(2)\\
& = 4.3.2.f(1)\\
& = 4.3.2.1.f(0)\\
& = 4.3.2.1.1\\
& = 24
\end{align}
$$

Penyelesaian perhitungan fungsi di atas dilakukan dengan cara terus memperkecil nilai input sebesar 1 satuan hingga masuk dalam kasus basis, yaitu $n=0$. Prinsip yang demikian juga dapat diterapkan untuk menyelesaikan berbagai permasalahan lain.

#### Topological Sorting

Dalam masalah *Topological Sorting*, kita akan diberikan serangkaian objek yang saling memiliki relasi yang bersifat antisimetris untuk kemudian 'diurutkan' berdasarkan restriksi relasi tersebut. Masalah ini merupakan salah satu yang paling bersifat *practical* karena terdapat banyak kejadian sehari-hari yang penyelesaiannya menggunakan pendekatan semacam demikian. 

Biasanya masalah ini sangat natural muncul dalam konteks manajemen proyek, yaitu ketika seorang manajer perlu untuk mengatur eksekusi dari banyak sekali pekerjaan dalam sebuah proyek, namun juga perlu memperhatikan bahwa pekerjaan tertentu tidak dapat dimulai jika pekerjaan yang lain belum selesai. Masalah *Topological Sorting* juga akan muncul secara natural ketika kita menghadapi objek-objek yang memiliki sifat keterurutan secara parsial. Objek matematika yang mendasari fenomena ini adalah [Partially Ordered Set (Poset)](https://en.wikipedia.org/wiki/Partially_ordered_set).

Struktur data yang lazim digunakan dalam penyelesaian masalah ini adalah graf berarah, di mana sebuah sisi berarah/*arc* antara dua buah titik menggambarkan sifat urutan dari kedua titik tersebut. Mungkin tidak semua pasangan titik dapat memiliki sisi berarah. Dalam kasus demikian, maka keterurutan kedua titik tersebut mungkin perlu diperoleh dengan membandingkan relasi keterurutan masing-masing titik dengan titik lain; atau mungkin juga memang kedua titik tersebut tidak memiliki hubungan keterurutan yang perlu diikuti secara ketat. 

Dengan demikian, ketika objek-objek diurutkan, mungkin hasil urutannya bisa saja tidak bersifat unik, mengingat adanya kemungkinan kelompok titik tertentu yang dapat muncul dengan urutan yang 'bebas'.

**Penyelesaian Topological Sorting**

Dengan strategi *Decrease and Conquer*, masalah ini dapat diselesaikan melalui penerapan lema berikut.

> **Lema** (Eksistensi titik sumber di sebuah DAG)
> 
> Misalkan $D=(V, A)$ merupakan sebuah DAG. Maka $D$ selalu memiliki sebuah **titik sumber**, yaitu titik yang tidak memiliki 'prasyarat' dari titik-titik lainnya. Yaitu, $$\exists v\in V, \quad \deg_{\text{in}}(v)=0.$$

> **Lema** (Penghapusan titik sumber tetap menghasilkan DAG)
> 
> Misalkan $D=(V, A)$ merupakan sebuah DAG dan $v$ adalah sebuah titik sumber. Maka, $D-\{v\}$ juga adalah DAG.

Perhatikan bahwa di sebuah DAG, titik sumber selalu dapat ditampilkan sebagai titik pertama dalam sebuah urutan topologis karena titik sumber tidak memiliki prasyarat dari titik manapun. Berdasarkan kedua lema di atas, prosedur di bawah ini akan menghasilkan urutan topologis.

```
Input: A directed graph D = (V, A)
Output: A topological sorting of the vertices in V, or print "ERROR" if such topological sorting does not exists.

TopologicalSorting(D):
  sorted = []
  while D is not empty:
    v, found = FindSourceVertex(D) # a procedure to find a source vertex
    if not found:
      return "ERROR"
    RemoveSourceVertex(D, v) # a procedure to remove vertex v from D
    sorted.append(v)
  
  return sorted
```

**Penyelesaian Subset Generation**

Kita dapat membangkitkan himpunan kuasa dari sebuah himpunan hingga dengan strategi *Decrease and Conquer*. Pertama, kita berfokus untuk membangkitkan seluruh string biner dengan panjang $N$.

Sebelumnya, perhatikan masalah hal ini sebetulnya dapat diselesaikan secara naif dengan cara mengiterasi seluruh bilangan bulat dari $0$ hingga $2^{N}-1$ dan secara naif mencari representasi bentuk biner dari masing-masing bilangan, kemudian mengumpulkannya.

Namun demikian, dalam kesempatan ini kita akan melihat penyelesaian yang lebih bersifat *Decrease and Conquer*. Pertama definisikan notasi-notasi berikut. Misalkan $A=\{a_0,\dots, a_{N-1}\}$. Untuk sembarang $i=0, 1, \dots, N-1$, definisikan $A_i=\{a_0, a_1, \dots, a_{i-1}\}$. Misalkan pula notasi $\mathcal{P}(A)$ menyatakan himpunan kuasa dari $A$. Maka, kita memiliki relasi rekursif sebagai berikut.

$$
\mathcal{P}(A_i) = \mathcal{P}(A_{i-1}) \cup \left\{\{a_{i-1}\}\cup B: B\in \mathcal{P}(A_{i-1})\right\},
$$

atau secara verbal, kita dapat memahami bahwa setiap himpunan kuasa dari $A_i$ dapat dibangun dari himpunan kuasa $A_{i-1}$, yaitu dengan 
1. Mengadopsi semua subhimpunan dalam himpunan kuasa $\mathcal{P}(A_{i-1})$ itu sendiri, dan
2. Menambahkan elemen $a_{i-1}$ (yang secara definisi pasti tidak termuat di $A_{i-1}$) pada setiap semua subhimpunan dalam himpunan kuasa $\mathcal{P}(A_{i-1})$.

Dalam bahasa string biner, algoritma untuk membangkitkan himpunan kuasa dapat dilakukan dengan membangkitkan semua kode biner yang mungkin. Berikut merupakan algoritma kasarnya.

```
GeneratePowerSet(A, i):
  if N==0:
    return an empty list
  
  # obtain a list of Power Set of A_{i-1} by GeneratePowerSet(i-1)
  L1 = GeneratePowerSet(i-1)
  # for each subset in L1, append the element a_{i-1} and add it to the power set
  L2 = []
  for subset in L1:
    modified_subset = subset appended by A[i-1]
    add modified_subset to L2
  
  return concatenation of L1 and L2
```

### 3.1.2 Decrease by a constant factor

Dalam strategi ini, ukuran input sebuah masalah akan berkurang proporsinya secara konstan dalam setiap iterasinya. Pendekatannya masih sama seperti upaya untuk menyelesaikan relasi rekurens, namun dengan 'jangkauan' langkah rekursif yang jauh lebih besar. Sekarang tinjau masalah menghitung pangkat bilangan $a^b$ untuk sembarang bilangan bulat nonnengatif $a, b$. Misalkan kita definisikan $f(a, b) = a^b$. Maka, fungsi tersebut dapat dihitung secara iteratif sebagai berikut.

$$
f(a, b)=
\begin{cases}
a f(a, b-1), & b>0\\
1, & b=0
\end{cases}
$$

Namun demikian, terdapat metode pemangkatan lain yang lebih efisien. Untuk contoh spesifik $a=2$ dan $b=10$, kita dapat menghitung $2^{20}$ sebagai berikut.
$$
\begin{align}
2^{20} &= (2^{10})^2\\
2^{10} &= (2^5)^2\\
2^{5} &= (2^2)^2 \times 2\\
2^{2} &= (2^1)^2 \\
2^{1} &= (2^0)^2 \times 2\\
\end{align}
$$
Jadi, daripada kita menghitung perkalian angka 2 sebanyak 20 kali, kita dapat menggunakan strategi di atas sehingga total operasi perkalian akan jauh lebih sedikit.

Untuk memperjelas hal ini, fungsi rekursif di atas dapat ditulis ulang dengan ekspresi yang berbeda sebagai berikut.


$$
f(a, b)=
\begin{cases}
f(a, \frac{b}{2})^2, & b>0 \text{ dan }b\text{ genap}\\
a\times f(a, \frac{b-1}{2})^2, & b>0 \text{ dan }b\text{ ganjil}\\
1, & b=0
\end{cases}
$$

Artinya, di setiap iterasi, ukuran input dari $f$ selalu berkurang secara proporsional, yaitu sekitar separuh kalinya. Hal ini jauh lebih cepat ketimbang di setiap iterasinya, penurunan ukuran input selalu bersifat konstan.

Dalam kesempatan ini kita akan melihat beberapa permasalahan yang dapat diselesaikan dengan strategi ini.

**Penyelesaian Masalah Koin Palsu**

Misalkan $A$ adalah himpunan yang berisi $N$ buah koin identik tersebut. Strategi *Decrease and Conquer* dari penyelesaian masalah ini adalah dengan membagi himpunan $A$ ke dalam dua kelompok besar dan menimbang kedua kelompok tersebut pada masing-masing tangan timbangan. Tentu kita mengharapkan bahwa banyaknya koin dari masing-masing tangan adalah sama. Oleh sebab itu, nantinya kita perlu memperhatikan secara khusus kasus di mana $A$ berjumlah ganjil.

Seandainya $A$ berisi sejumlah genap koin, maka kita dapat membagi $A$ ke dalam dua bagian sama besar, katakanlah $A_1$ dan $A_2$. Maka timbang kedua kelompok tersebut. Kelompok yang terbukti lebih ringan dibanding kelompok yang lain pasti memuat koin palsu tersebut.

Selanjutnya, jika $A$ berisi sejumlah ganjil koin, maka pertama sisihkan sebuah koin (bebas) terlebih dahulu, kemudian, sisanya dibagi ke dalam dua bagian sama besar, katakanlah $A_1$ dan $A_2$. Maka timbang kedua kelompok tersebut. Kelompok yang terbukti lebih ringan dibanding kelompok yang lain pasti memuat koin palsu tersebut. Namun jika ternyata kedua kelompok sama berat, maka kita sebetulnya sudah menemukan bahwa koin yang ditimbanglah yang merupakan koin palsu.

Maka, dengan sekali penimbangan melalui strategi di atas, kita dapat memperkecil ukuran pencarian kita menjadi sekitar separuhnya. Pada iterasi berikutnya, algoritma yang persis sama dapat diterapkan kembali hingga akhirnya ukuran koin cukup kecil sehingga koin palsu dapat ditemukan dengan mudah.

Berikut merupakan algoritma kasar penyelesaian di atas.

```
FakeCoin(A):
  if size(A)==1:
    return A

  if size(A) is even:
    # divide A into two classes of equal size
    A1, A2 = PartitionIntoTwoClasses(A)
    
    # perform the scaling operation, the lighter group is denoted by A_decreased
    A_decreased = scale(A1, A2)
    return FakeCoin(A_decreased)

  if  size(A) is odd:
    # divide A into three classes: A1 & A2 of equal size and a single coin c
    A1, A2, c = PartitionIntoThreeClasses(A)

    # perform the scaling operation
    if scale(A1, A2) is balanced:
      return c
    else:
      A_decreased = scale(A1, A2)
      return FakeCoin(A_decreased)
```

## 3.2 Divide and Conquer

Strategi ini mirip dengan *Decrease and Conquer* dalam hal permasalahan akan dibawa ke dalam bentuk yang lebih kecil ukurannya. Namun demikian, dalam strategi *Divide and Conquer*, biasanya memiliki tambahan operasi, yaitu menggabungkan kembali hasi penyelesaian subpermasalahan yang lebih kecil tersebut.

Berikut merupakan kerangka umum strategi *Divide and Conquer*:
1. Input permasalahan dipecah ke dalam sejumlah subpermasalahan yang lebih kecil, biasanya masing-masing memiliki ukuran yang kurang lebih sama.
2. Setiap permasalahan kembali diselesaikan dengan strategi yang sama secara rekursif, hingga diperoleh ukuran input yang cukup kecil untuk dapat diselesaikan dengan 'mudah'.
3. Setiap penyelesaian dari subpermasalahan digabungkan kembali untuk menghasilkan solusi dari permasalahan utama.

### 3.2.1 Master Theorem

Dalam menganalisis algoritma dengan strategi *Divide and Conquer*, biasanya kita akan menemui persamaan rekurens yang 'standar' sebagai berikut. Misalkan $T(N)$ adalah banyaknya operasi dasar yang dilakukan oleh algoritma untuk masalah dengan ukuran $N$. Maka,

$$
T(N) =aT\left(\dfrac{N}{b}\right) + f(N),
$$
dengan penjelasan sebagai berikut:

* Strategi *Divide and Conquer* membagi permasalahan berukuran $N$ menjadi $a$ buah subpermasalahan yang lebih kecil, yang masing-masing berukuran $\frac{N}{b}$, untuk suatu bilangan bulat positif $a$ dan $b$. Banyaknya operasi yang dilakukan adalah sebesar $aT\left(\frac{N}{b}\right)$.
* Jika diperlukan, mungkin dibutuhkan tambahan operasi yang dilakukan untuk membagi permasalahan ke dalam subpermasalahan dan juga untuk menggabungkan kembali hasil penyelesaian subpermasalahan untuk menjadi solusi yang utuh. Banyaknya operasi yang dilakukan adalah sebesar $f(N)$.

Terdapat sebuah teorema yang dapat secara tuntas menjawab berapa besar kompleksitas $T(n)$ secara asimtotik di mana hasil akhirnya bergantung pada nilai $a$, $b$, dan $f(n)$.

> **Master Theorem**
> 
> Misalkan fungsi $T(n)$ memenuhi persamaan rekursif $$T(N) =aT\left(\dfrac{N}{b}\right) + f(N)$$ dengan $f(N)= \Theta(n^d)$ untuk suatu $d\ge 0$. Maka, 
$$
T(N)=\begin{cases}
\Theta(n^d), & \text{jika }a<b^d\\
\Theta(n^d\log n), & \text{jika }a=b^d\\
\Theta(n^{\log_b a}), & \text{jika }a>b^d\\
\end{cases}
$$

### 3.2.4 Closest Pair with Divide and Conquer

Permasalahan mencari pasangan titik terpendek di antara $N$ titik tersedia dapat diselesaikan secara naif dengan kompleksitas waktu $\mathcal{O}(N^2)$.

Dalam bagian ini, kita akan melihat bagaimana permasalahan ini dapat diselesaikan secara Divide and Conquer sehingga akhirnya dapat memiliki kompleksitas waktu yang lebih efisien.

Ide utama dalam strategi ini adalah dengan membagi $N$ buah titik yang hendak dicari pasangan terdekatnya ke dalam dua kelompok berdasarkan posisi koordinat $x$ nya. Kemudian masing-masing kelompok dicari pasangan terdekatnya dengan algoritma yang sama secara rekursif untuk nantinya dibandingkan mana di antaranya yang lebih kecil.

Berikut merupakan algoritma kasarnya:

```
Misalkan list L menyimpan N buah titik (x_i, y_i) yang sudah diurutkan berdasarkan nilai x_i nya.

ClosestPairDnC(L):
  n = len(L)
  if n <= 3:
    solve by exhaustive search
  else
    copy all elements in L[0:n/2] to L1
    copy all elements in L[n/2:n] to L2

    d1 = ClosestPairDnC(L1)
    d2 = ClosestPairDnC(L2)
    d = min(d1, d2)

    gather all elements in L whose x coordinates is <= d into the list P
    sort P by the y coordinates

    search within P the pairs whose distance < d, then updates the d accordingly
    # make use the sorted values of P to get an efficient search

    return d
```
